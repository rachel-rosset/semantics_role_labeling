```python
from decomp import UDSCorpus

uds = UDSCorpus()

print(uds["ewt-train-12"])
```

```python

uds_train = UDSCorpus(split='train')
print("done")
```

```python
querystr = """
           SELECT ?pred
           WHERE { ?pred <domain> <semantics> ;
                         <type> <predicate> ;
                         <factual> ?factual ;
                         <dur-minutes> ?duration
                         FILTER ( ?factual > 0 && ?duration > 0 )
                 }
           """

results = {gid: graph.query(querystr, query_type='node', cache_rdf=False)
           for gid, graph in uds.items()}
print("done")
```

make new dictionary of 1000 items from uds train dict

```python
short_train = {k: uds_train[k] for k in list(uds_train)[:1000]}

print(len(short_train))
print(short_train)
```

```python
agentstr = """
           SELECT ?edge
           WHERE { ?pred ?edge ?arg ;
                         <domain> <semantics> ;
                         <type>   <predicate> .
                   ?arg  <domain> <semantics> ;
                         <type>   <argument>  .
                    
                    {
                       ?edge <volition> ?volition
                         FILTER ( ?volition > 0 ) 
                    }
                        UNION
                    {
                        ?edge <instigation> ?instigation
                        FILTER ( ?instigation > 0) 
                    }
                    ?edge <existed-before> ?existed-before
                        FILTER ( ?existed-before > 0)
                 }
           """

#results = short_train.query(querystr, query_type='node', cache_rdf=False)
results = {gid: graph.query(agentstr, query_type='edge', cache_rdf=False)
           for gid, graph in short_train.items()}
print("done")
```

Agent Query

```python
agentstr = """
           SELECT ?edge
           WHERE { ?pred ?edge ?arg ;
                         <domain> <semantics> ;
                         <type>   <predicate>.
                   ?arg  <domain> <semantics> ;
                         <type>   <argument>  .
                    ?edge <existed_before> ?existed_before
                        FILTER ( ?existed_before > 0) .
                    { ?edge <volition> ?volition
                         FILTER ( ?volition > 0 ) 
                    } UNION
                    { ?edge <instigation> ?instigation
                        FILTER ( ?instigation > 0) 
                    }
                    
                 }
           """

results = {gid: graph.query(agentstr, query_type='edge', cache_rdf=False)
           for gid, graph in short_train.items()}


print("done")
```

```python
for r in results.items():
    print(r)
```

Patient Query

```python
patient_str = """
           SELECT ?edge
           WHERE { ?pred ?edge ?arg ;
                         <domain> <semantics> ;
                         <type>   <predicate> .
                   ?arg  <domain> <semantics> ;
                         <type>   <argument>  .
                    ?edge <existed_before> ?existed_before
                        FILTER ( ?existed_before > 0) .
                    { ?edge <change_of_state> ?change_of_state
                         FILTER ( ?change_of_state > 0 ) 
                    } UNION
                    { ?edge <change_of_state_continuous> ?change_of_state_continuous
                        FILTER ( ?change_of_state_continuous > 0) 
                    }
                    
                 }
           """

patient_results_extra = {gid: short_train[gid].query(patient_str, query_type='edge', cache_rdf=False)
           for gid in short_train}


print("done")
```

```python
patient_results = {}
# new dictionary with empty key,value pairs removed 
for documentid, document in patient_results_extra.items():
    if document != {}:
        patient_results[documentid] = document

for documentid, document in patient_results.items():
    print(short_train[documentid].sentence)
    # syntax nodes gives tokenized sentence (short_train[documentid].syntax_nodes)
    #print(short_train[documentid].syntax_nodes)
    #semantics edges(None, "head") gives the head
    print(short_train[documentid].semantics_edges(None,"head"))
    print(len(short_train[documentid].semantics_edges(None,"head")))
    print("doc id", document)
    print("----------")
    #rint(document)
    
print(len(patient_results.keys()))

# short_train[documentid] to get the sentence graph representation
# patient_results holds a dictionary mapping documentid's to edges with certain attributes
```

Instrument Query

```python
instrument_str = """
           SELECT ?edge
           WHERE { ?pred ?edge ?arg ;
                         <domain> <semantics> ;
                         <type>   <predicate> ;
                         <pred-particular> ?ppart
                         FILTER ( ?ppart > 0 ) .
                   ?arg  <domain> <semantics> ;
                         <type>   <argument>  ;
                         <arg-particular> ?apart
                         FILTER ( ?apart > 0 ) .
                    ?edge <existed_during> ?existed_during
                        FILTER ( ?existed_during > 0) .
                    ?edge <was_used> ?was_used
                         FILTER ( ?was_used > 0 ) .
                    
                 }
           """

instrument_results = {gid: graph.query(instrument_str, query_type='edge', cache_rdf=False)
           for gid, graph in short_train.items()}


print("done")
```
